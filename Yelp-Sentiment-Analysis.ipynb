# Getting Reviews from all pages

```python
num_pages = 20
base_url = 'https://www.yelp.com/biz/dishoom-london?osq=Restaurants&start='
review_list=[]

for i in range(num_pages):
        url = f'{base_url}{i*20}'
        r = requests.get(url)
        soup=BeautifulSoup(r.text,'html.parser')
        divs=soup.find_all(class_='review__09f24__oHr9V')
        
        for div in divs:
            review=(div.find('p').text)
                    
            review_list.append(review)
        
        
        
```

# Analysing the data

```python
import pandas as pd
import numpy as np
import nltk

```

```python
df=pd.DataFrame(np.array(review_list), columns=['review'])
df.head()
```

```python
len(df.review)
```

```python
df['word_count']=df['review'].apply(lambda x: len(x.split()))
df['char_count']=df['review'].apply(lambda x: len(x))
```

```python
def average_words(x):
    words = x.split()
    return sum(len(word) for word in words) / len(words)
df['average_word_length']=df['review'].apply(lambda x: average_words(x))
```

```python
df['stopwords_count']=df['review'].apply(lambda x: len(list(word for word in x.split() if word.lower() in stop_words)))
```

```python
df['lowercase']=df['review'].apply(lambda x:  " ".join(word.lower() for word in x.split()))
df['punctuations']=df['lowercase'].str.replace('[^\w\s]','')
```

```python
df['stopwords']=df['punctuations'].apply(lambda x:  " ".join(word for word in x.split() if word not in stop_words))
```

# Sentiment Analysis

```python
import textblob
```

```python
from textblob import Word
df['lemmatized']=df['stopwords'].apply(lambda x: " ".join(Word(word).lemmatize() for word in x.split()))
```

```python
from textblob import TextBlob
df['polarity']=df['lemmatized'].apply(lambda x: TextBlob(x).sentiment[0])
df['subjectivity']=df['lemmatized'].apply(lambda x: TextBlob(x).sentiment[1])
df.drop(['lowercase','punctuations','stopwords'] ,axis=1,inplace=True)
df
```
